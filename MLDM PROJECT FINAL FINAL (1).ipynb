###  Extraction of dependent and independent variables

```python
#Extraction of dependent and independent variables
X=data.iloc[:,1:] #Independent Var
Y=data.iloc[:,0:1] #dependent var
```

#### Handling the categorical data

```python
X=pd.get_dummies(X)
X
```

## 3. Splitting the dataset into train (80%) and test (20%)

```python
#Split the data set into train 80% and test 20%
# Independent var: Present_Price, Kms_Driven,Fuel_Type,Transmission  
# Dependent var: Selling price 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2)
```

## 4. Feature Scaling

```python
#Feature scaling--Standrdization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

------------------

# Model Creation

#  Multiple Linear Regression
##### The model of MLR is as Follows:  
##### y=β1x1+β2x2+⋯+βnxn+β0
       ### here Y is the target variable
       ### X1,X2,Xn... is the dependent  variables
       ### Beta0 is the intercept
       ### beta1 is the coefficient of Regression

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_squared_log_error
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,Y_train)
y_pred_3 = LR.predict(X_test)
```

```python
print("r2 score is ---->:",r2_score(Y_test,y_pred_3)*100)
print("mae is ---->:",mean_absolute_error(Y_test,y_pred_3))
print("MSQ is ---->:",mean_squared_error(Y_test,y_pred_3))
```

```python
plt.scatter(Y_test,y_pred_3)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## KNN Regressor

```python
# import Lib
from sklearn.neighbors import KNeighborsRegressor
```

```python
#estimating MSE for k=1-9
mse=[]
for i in range(1,10):
    KNN=KNeighborsRegressor(n_neighbors=i)
    KNN.fit(X_train,Y_train)
    y_pred=KNN.predict(X_test)
    error=np.sqrt(mean_squared_log_error(Y_test, y_pred))
    mse.append(error)
    print("K =",i," , Root mse =",error)
```

```python
#plotting error

graph_error = pd.DataFrame(mse) #elbow curve 
plt.plot(list(range(1,10)),mse)
plt.xlabel('K')
plt.ylabel('MSE')
plt.title('Elbow Curve')
plt.show()
```

### <Observaion:> 
__From this graph we can observe that the lowest bend is at k=6 thereofore indicates appropriate number of clusters and also we could observe that at k=5 error is least and r2 square is maximum i.e., 80.07%.__
